```
import requests
from bs4 import BeautifulSoup
```

```
webdata=requests.get("https://en.wikipedia.org/wiki/Machine_learning")
webdata
```

```
fulldata=webdata.text
print(fulldata)
```

```
soup=BeautifulSoup(fulldata,'html5lib')
#print(soup)
```

```
parsedata=soup.get_text()
print(parsedata)
```

```
with open('mlwiki.txt','w+') as f:
  f.write(parsedata)
```

```
f=open('mlwiki.txt','r')
readdata=f.read()
f.close()

#readdata
```

```
newdata=[i for i in readdata.split()] 
#split means removing spaces bw words
print(newdata)
```

```
import nltk 
counteddata=nltk.FreqDist(newdata)
#counteddata
counteddata.plot(20)
```

```
## 2 TOKENIZATION
import nltk
nltk.download("punkt")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
tokdata=word_tokenize(parsedata)
```

```
# 3 STOPWORDS 
tok=[i for i in tokdata if i.lower() not in stopwords.words('english')] 
```

```
tok
```

```
## 4 STEMMING
from nltk import stem
from nltk.stem import PorterStemmer


```

```
stemfind=PorterStemmer()
```

```
word=stemfind.stem('learning')
```

```
## REPLACING
[i.replace(word,"LEARN") for i in tok ]
```
