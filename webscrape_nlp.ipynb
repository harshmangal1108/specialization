```
[i for i in dir(nltk) if 'Freq' in i]
```

```
nlpdata=nltk.FreqDist(newdata)
nlpdata.plot()
```

```
### REMOVING STOPWORDS
removedata=[i for i in newdata if i.lower() not in stopwords.words('english')]
```

```
## AGAIN APPLY NLTK.FREQDIST
filterdata=nltk.FreqDist(removedata)
```
